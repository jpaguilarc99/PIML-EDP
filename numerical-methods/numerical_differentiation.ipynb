## Finite Differences Method

Encontrar la primera y segunda derivada del polinomio en $x=0.1$

- $f(x) = 0.1x^5 - 0.2x^3 + 0.1x - 0.2$

La diferenciacion analitica produce

- $f'(x) = 0.5x^4 - 0.6x^2 + 0.1$
- $f''(x) = 2x^3 - 1.2x$

Sustituyendo $x = 0.1$

$f'(0.1) = 0.09405$

$f''(0.1) = -0.118$

```python
# Forward finite difference approximation
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
h = 0.01
x = 0.1

dff1 = (f(x+h) - f(x)) / h
dff2 = (f(x+2*h) - 2*f(x+h) + f(x)) / h**2

print('f\'(0.1) =', dff1)
print('f\'\'(0.1) =', dff2)
```

```python
# Central finite difference approximation
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
h = 0.01
x = 0.1

dfc1 = (f(x+h) - f(x-h)) / (2*h)
dfc2 = (f(x+h) - 2*f(x) + f(x-h)) / h**2

print('f\'(0.1) =', dfc1)
print('f\'\'(0.1) =', dfc2)
```

```python
# Backward finite difference approximation
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
h = 0.01
x = 0.1

dfb1 = (f(x) - f(x-h)) / h
dfb2 = (f(x) - 2*f(x-h) + f(x-2*h)) / h**2

print('f\'(0.1) =', dfb1)
print('f\'\'(0.1) =', dfb2)
```

### Plotting Derivative Curves

```python
# Central finite difference approximation
import numpy as np
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
h = 0.01
x = np.linspace(0, 1, 11)
print(f"Domain {x}\n")

dfc1 = (f(x+h) - f(x-h)) / (2*h)
dfc2 = (f(x+h) - 2*f(x) + f(x-h)) / h**2
print(f"dfc1 -> {dfc1}\n")
print(f"dfc2 -> {dfc2}")
```

```python
import matplotlib.pyplot as plt
plt.plot(x, f(x), '-k', x, dfc1, '--b', x, dfc2, '-.r')
plt.xlabel('x')
plt.ylabel('y, y\', y\'\'')
plt.legend(['y', 'y\'', 'y\'\''])
plt.grid()
plt.show()
```

### Differentiation Scipy

```python
from scipy.misc import derivative
f = lambda x: 0.1*x**5 - 0.2*x**3 + 0.1*x - 0.2
derivative(f, 0.1, 0.01, 1) # 1st derivative
```

```python
derivative(f, 0.1, 0.01, 2) # 2nd derivative
```
